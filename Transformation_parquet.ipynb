```python
import pandas as pd
import duckdb 
import geopandas as gpd
import geoparquet_io as gpio
from shapely.ops import transform
```

```python
df = gpd.read_file('france_metro/carreaux_200m_met.shp')
```

```python
df = df.to_crs("EPSG:3035")
```

```python
def round_geometry(geom, precision=3):
    return transform(lambda x, y, z=None: (round(x, precision), round(y, precision)), geom)
```

```python
df["geometry"] = df["geometry"].apply(lambda g: round_geometry(g, 0))
```

```python
df.drop(columns=['idcar_1km', 'idcar_nat', 'i_est_200', 'i_est_1km',
       'lcog_geo']).to_parquet('carreaux_200m_met.parquet', index=False)
```

```python
gpio.read('carreaux_200m_met.parquet') \
    .add_bbox() \
    .sort_hilbert() \
    .write('carreaux-200m-met-3035_2021.parquet', compression='ZSTD')
```
